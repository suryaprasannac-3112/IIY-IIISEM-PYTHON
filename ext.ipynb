```python
def fibonacci(n):
  if n<=0:
    print("invalid input")
  elif n==1:
    return 0
  elif n==2:
    return 1
  else:
    a,b=0,1
    for _ in range(2,n):
      a,b = b,a+b
    return b
n = int(input("Enter the value of n: "))
result = fibonacci(n)
print(f"The {n}th Fibonacci number is: {result}")
```

```python
class Animal:
  def __init__(self,name):
    self.name=name
  def eat(self):
    print(f"{self.name} is eating")
  def sleep(self):
    print(f"{self.name} is sleeping")
class Dog(Animal):
  def bark(self):
    print(f"{self.name} is barking")
class Cat(Animal):
  def meow(self):
    print(f"{self.name} is saying meow")
d1=Dog("buddy")
c1=Cat("tom")
d1.eat()
d1.sleep()
c1.eat()
c1.sleep()
d1.bark()
c1.meow()
```

```python
import numpy as np
arr1=np.linspace(1,10,5)
arr2=np.ones(5) * 5
print("Array 1 : ", arr1)
print("Array 2 : ",arr2)
```

```python
max_value=np.max(arr1)
min_value=np.min(arr1)
max_index=np.argmax(arr1)
min_index=np.argmin(arr1)
print("Max Value : ",max_value)
print("Min Value : ",min_value)
print("Max Index : ",max_index)
print("Min Index : ",min_index)
```

```python
import numpy as np

# Generate array of 20 random integers between 1 and 100
arr = np.random.randint(1, 101, 20)

print("Generated Array:\n", arr)
```

```python
# Using np.unique to find unique elements and their counts
unique_elements, counts = np.unique(arr, return_counts=True)

print("\nUnique Elements:\n", unique_elements)
print("Counts of Each Element:\n", counts)
```

```python
import pandas as pd
df=pd.DataFrame(arr,columns=['Numbers'])
mean_value=df['Numbers'].mean()
median_value=df['Numbers'].median()
standard_deviation=df['Numbers'].std()
print("Mean : ",mean_value)
print("Median : ",median_value)
print("Standard Deviation : ",standard_deviation)
```

```python
import matplotlib.pyplot as plt
plt.hist(df['Numbers'],edgecolor='black')
plt.xlabel('values')
plt.ylabel('frequency')
plt.title('Histogram of Numbers')
plt.show()
```
